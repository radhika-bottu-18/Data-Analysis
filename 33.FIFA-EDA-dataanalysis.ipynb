```python
def sinplot(flip=1):
    x=np.linspace(1,14,100)
    for i in range(1,7):
        plt.plot(x,np.sin(x+i*.5)*(7-i)*flip)
```

```python
sinplot()
```

```python
def cosplot(flip=1):
    x= np.linspace(1,14,100)
    for i in range(0,7):
        plt.plot(x,np.cos(x*i+0.5)*(7-i)*flip)
```

```python
cosplot()
```
